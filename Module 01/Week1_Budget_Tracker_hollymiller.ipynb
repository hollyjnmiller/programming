### Paste and Test Your AI-Generated Budget Tracker

````python
# AI-GENERATED BUDGET TRACKER CODE
# [Paste your code here]

monthly_income = float(input("Enter your monthly income: "))

expenses = []
for i in range(1, 4):
    while True:
        try:
            expense = float(input(f"Enter expense {i}: "))
            expenses.append(expense)
            break
        except ValueError:
            print("Invalid input. Please enter a number for the expense.")

total_expenses = sum(expenses)
remaining_budget = monthly_income - total_expenses

print(f"\nRemaining budget: ${remaining_budget:.2f}")

if remaining_budget < 0:
    print(f"Warning: You are over budget by ${abs(remaining_budget):.2f}!")
````

## Part 4: Enhancement Challenge

Pick ONE feature to add with AI's help:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 A detailed budget summary with advice


````python
# Document your enhancement choice and the prompt you used
enhancement_choice = "I chose to add: Expense categories (food, transport, entertainment, other)"
enhancement_prompt = """
My Enhanced Budget Tracker Prompt (based on previous code):

I have the following Python code for a budget tracker:

```python
monthly_income = float(input("Enter your monthly income: "))

expenses = []
for i in range(1, 4):
    while True:
        try:
            expense = float(input(f"Enter expense {i}: "))
            expenses.append(expense)
            break
        except ValueError:
            print("Invalid input. Please enter a number for the expense.")

total_expenses = sum(expenses)
remaining_budget = monthly_income - total_expenses

print(f"\nRemaining budget: ${remaining_budget:.2f}")

if remaining_budget < 0:
    print(f"Warning: You are over budget by ${abs(remaining_budget):.2f}!")
```

Modify this code to add the following features:
- For each of the up to 3 expenses, after asking for the expense amount, also ask the user to enter a category for that expense (e.g., 'food', 'transport', 'entertainment', 'other').
- Store both the expense amount and its category.
- Calculate the total amount spent for each category.
- In the final output, in addition to the remaining budget and the over-budget warning, print a summary that shows the total amount spent in each category.
"""

print(f"🚀 Enhancement: {enhancement_choice}")
````

````python
# ENHANCED BUDGET TRACKER
# AI-GENERATED CODE

monthly_income = float(input("Enter your monthly income: "))

expenses_with_categories = []
for i in range(1, 4):
    while True:
        try:
            expense_amount = float(input(f"Enter amount for expense {i}: "))
            expense_category = input(f"Enter category for expense {i} (e.g., food, transport, entertainment, other): ").lower()
            expenses_with_categories.append({'amount': expense_amount, 'category': expense_category})
            break
        except ValueError:
            print("Invalid input. Please enter a number for the expense amount.")

total_expenses = sum(item['amount'] for item in expenses_with_categories)
remaining_budget = monthly_income - total_expenses

print(f"\nRemaining budget: ${remaining_budget:.2f}")

if remaining_budget < 0:
    print(f"Warning: You are over budget by ${abs(remaining_budget):.2f}!")

# Calculate totals per category
category_totals = {}
for item in expenses_with_categories:
    category = item['category']
    amount = item['amount']
    category_totals[category] = category_totals.get(category, 0) + amount

print("\n--- Expense Summary by Category ---")
if category_totals:
    for category, total in category_totals.items():
        print(f"{category.capitalize()}: ${total:.2f}")
else:
    print("No expenses entered.")
````
